## Consulta 2

Cantidad de películas por plataforma con un puntaje mayor a XX en determinado año.

La función debe llamarse get_score_count(platform, scored, year)

```python
# 1. Filtro por año
anio = 2015
filtro_anio = data_plataformas.loc[ data_plataformas['release_year'] == anio ]
filtro_anio
```

```python
# 2. Filtro por score
score = 3.5    
filtro_score = filtro_anio.loc[ filtro_anio['score'] > score ]
filtro_score
```

```python
# 3. Agrupar por plataforma y contar peliculas

#creamos la columna 'platform' para poder agrupar más adelante
filtro_score['plataforma'] = filtro_score['Id'].str[0]

# Agrupamos las películas por plataforma y contamos el número de películas en cada plataforma
peliculas_x_plataformas = filtro_score.groupby('plataforma').size().reset_index(name='count')
peliculas_x_plataformas
```

```python
# 4. Filtrar por plataforma

plataforma = 'a'
plataforma_y_peliculas = peliculas_x_plataformas.loc[ peliculas_x_plataformas['plataforma'] == plataforma ]
plataforma_y_peliculas
```

```python
# Todo junto

# 1. Filtro por año
anio = 2015
filtro_anio = data_plataformas.loc[ data_plataformas['release_year'] == anio ]

# 2. Filtro por score
score = 3.5    
filtro_score = filtro_anio.loc[ filtro_anio['score'] > score ]

# 3. Agrupar por plataforma y contar peliculas
  #creamos la columna 'platform' para poder agrupar más adelante
filtro_score['plataforma'] = filtro_score['Id'].str[0]

  # Agrupamos las películas por plataforma y contamos el número de películas en cada plataforma
peliculas_x_plataformas = filtro_score.groupby('plataforma').size().reset_index(name='count')

# 4. Filtrar por plataforma
plataforma = 'a'
plataforma_y_peliculas = peliculas_x_plataformas.loc[ peliculas_x_plataformas['plataforma'] == plataforma ]

resultado = plataforma_y_peliculas [ plataforma_y_peliculas['plataforma'] == plataforma[0] ]
resultado = resultado['count']

print('La plataforma', plataforma, 'tiene', int(resultado), 'peliculas, publicadas en el año', anio, 'con score mayor a', score)
```
